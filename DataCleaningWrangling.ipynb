Using IP Address to get Region and City details

```python
import requests

def get_location(ip_address):
    try:
        response = requests.get(f"https://ipapi.co/{ip_address}/json/")
        data = response.json()

        city = data.get('city', '')
        region = data.get('region', '')

        return city, region
    except Exception as e:
        print(f"An error occurred while processing IP {ip_address}: {e}")
    return '', ''

for index, row in df_2.iterrows():
    if pd.isna(row['City / Town / District']) or pd.isna(row['Region']):
        ip_address = row['IP Address']
        if pd.notna(ip_address):
            city, region = get_location(ip_address)
            df_2.at[index, 'City / Town / District'] = city
            df_2.at[index, 'Region'] = region
        else:
            print(f"Invalid IP address at index {index}: IP={ip_address}")
```

Now we will fix the 'Country of Citizenship' column, capitalizing the first letter of the countries and making the list uniform.



```python
country_mapping = {
    'US': 'United States',
    'USA': 'United States',
    'Usa': 'United States',
    'United States ': 'United States',
    'United States, Pakistan': 'United States, Pakistan',
    'pakistan': 'Pakistan',
    'china': 'China'
}

df_2['Country of Citizenship'] = df_2['Country of Citizenship'].str.strip()
df_2['Country of Citizenship'] = df_2['Country of Citizenship'].replace(country_mapping)
df_2['Country of Citizenship'] = df_2['Country of Citizenship'].str.title()

```

Filling the missing values in 'Country of Citizenship' with the mode as the categorical nature of this column makes mode the most appropriate method to fill values.

```python
mode = df_2['Country of Citizenship'].mode()[0]
df_2['Country of Citizenship'].fillna(mode, inplace=True)
```

Now I will focus on the 'Height (Inches)' column and fix the erronous values such as 5.2, 6i and 7O. I will also use the mean height values to fill the missing value since it follows the normal distribution.


```python
df_2['Height (Inches)'] = df_2['Height (Inches)'].str.replace('5.2', '52')
df_2['Height (Inches)'] = df_2['Height (Inches)'].str.replace('6l', '66')
df_2['Height (Inches)'] = df_2['Height (Inches)'].str.replace('7O', '77')

df_2['Height (Inches)'] = pd.to_numeric(df_2['Height (Inches)'], errors='coerce')
df_2['Height (Inches)'].fillna(df_2['Height (Inches)'].mean(), inplace=True)
df_2['Height (Inches)'] = df_2['Height (Inches)'].astype(int)
```

```python
degree_mapping = {
    'Economic': 'Economics',
    'Economics': 'Economics',
    'Economics With Finance': 'Economics',
    'Economicss With Finance': 'Economics',

    'Finance': 'Finance',
    'Accounting': 'Finance',
    'Bs Accounting': 'Finance',
    'Finance And Isom': 'Finance',
    'Accounting And Isom': 'Finance',
    'Mathematics & Finance': 'Finance',
    'Bs Finance' : 'Finance',


    'Computer Science': 'Computer Science',
    'Computer science': 'Computer Science',
    'Computer Applications': 'Computer Science',
    'Computer Engineering': 'Computer Science',
    'Cse': 'Computer Science',
    'Bit': 'Computer Science',
    'B.Tech In Cs': 'Computer Science',
    'B.Tech In Computer Science': 'Computer Science',
    'Bachelor Of Technology In Computer Science': 'Computer Science',

    'Engineering': 'Engineering',
    'Electrical And Computer Engineering': 'Engineering',
    'Civil Engineering': 'Engineering',
    'Bachelor Of Engineering': 'Engineering',
    'Systems Engineering': 'Engineering',
    'Ece': 'Engineering',

    'Business Management': 'Business',
    'Management': 'Business',

    'Biology': 'Biology',
    'Psychology': 'Psychology',
    'Mathematics & Statistics': 'Mathematics & Statistics',
    'Quantity Surveying': 'Surveying',
    'Surveying': 'Surveying',
    'Criminology Law And Soci': 'Criminology, Law, and Sociology',
    'Criminology, Law And Sociology': 'Criminology, Law, and Sociology',
    'Unknown': 'Unknown',
    'Yes': 'Unknown'
}

df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].str.strip()
df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].replace(degree_mapping, regex=True)
df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].str.title()
df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].str.replace('Economicss', 'Economics')
df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].str.replace('B.Tech In Cs', 'Computer Science')
df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].str.replace('B.Tech In Computer Science', 'Computer Science')
df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].str.replace('Bachelor Of Engineering', 'Engineering')
df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].str.replace('Civil Engineering', 'Engineering')
df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].str.replace('Ece', 'Engineering')

df_2['Undergraduate Degree'].fillna('Unknown', inplace=True)
mode_degree = df_2['Undergraduate Degree'].mode()[0]
df_2['Undergraduate Degree'] = df_2['Undergraduate Degree'].replace('Unknown', mode_degree)

print(df_2['Undergraduate Degree'])
```
